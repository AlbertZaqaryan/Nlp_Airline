```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.tsv', delimiter='\t')
df_test = pd.read_csv('test.tsv', delimiter='\t')
df_sub = pd.read_csv('sample_submission.csv')
```

```python
df_train = df_train.dropna() 
```

```python
df_train
```

```python
y = df_train['sentiment']
X = df_train['verified_reviews']
```

```python
df_test = df_test['verified_reviews']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
X = tfidf.fit_transform(X)
```

```python
df_test = tfidf.transform(df_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
param = {
    'criterion':["gini", "entropy", "log_loss"],
    'min_samples_leaf':[1, 2, 3],
    'min_samples_split':[2, 3, 5, 7],
    'min_weight_fraction_leaf':[0.0, 0.5, 0.7]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
full_model = GridSearchCV(model, param_grid=param, verbose=2, cv=3, scoring='accuracy')
```

```python
full_model.fit(X,y)
```

```python
y_pred = full_model.predict(df_test)
```

```python
df_sub['sentiment'] = y_pred
```
